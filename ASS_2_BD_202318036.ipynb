```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
stop_words = set(stopwords.words('english'))
```

```python
def get_tokenized_list(doc_text):
    tokens = nltk.word_tokenize(doc_text)
    return tokens
print(get_tokenized_list("That would be too Painful!)"))

def word_stemmer(token_list):
  porter_stemmer = nltk.stem.PorterStemmer()
  stemmed = []
  for words in token_list:
    stemmed.append(porter_stemmer.stem(words))
  return stemmed
print(word_stemmer("That would be too Painful!)"))
```

```python
def remove_stopwords(doc_text):
  final_txt = []
  for words in doc_text:
    if words not in stop_words:
      final_txt.append(words)
  return final_txt
```

```python
tokens = get_tokenized_list(news_grp.data[0])
print(tokens)
```

```python
txt = remove_stopwords(tokens)
txt
```

```python

txt = word_stemmer(txt)#after stemming
txt
```

```python
doc_ = ' '.join(txt)
doc_
```

```python
final_corpus = []
for doc in news_grp.data:
  tokens = get_tokenized_list(doc)
  txt = remove_stopwords(tokens)
  txt  = word_stemmer(txt)
  txt = ' '.join(txt)
  final_corpus.append(txt)
final_corpus
```

```python
vectorizerX = TfidfVectorizer()
vectorizerX.fit(final_corpus)
doc_vector = vectorizerX.transform(final_corpus)
print(vectorizerX.get_feature_names_out())

print(doc_vector.shape)
```

```python
data2 = pd.DataFrame(doc_vector.toarray(), columns=vectorizerX.get_feature_names_out())
data2
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def cosi_similar(doc_vector, query_vector):
  cosineSimilarities = cosine_similarity(doc_vector,query_vector).flatten()
  return cosineSimilarities, len(cosineSimilarities), max(cosineSimilarities)

cosSim, l, maxSim = cosi_similar(doc_vector, x_vector)
cosSim
```
