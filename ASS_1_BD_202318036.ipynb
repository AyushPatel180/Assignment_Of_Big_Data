```python
#Q1 Block Matrix Multiplication

import numpy as np

def block(A,B):
	n1=len(A)
	n2=len(B[0])
	n3=len(A[0])
	C=[]
	for i in range(n1):
		list1=[]
		for j in range(n2):
			result=0
			for k in range(n3):
				result=result+ A[i][k]*B[k][j]
			list1.append(result)
		C.append(list1)
	return np.array(C)


def sub_mat(A,B):

  n = len(A)
  if n <= 2:
    return block(A, B)

  sub_mat_len_A=len(A)//2


  A1=A[0:sub_mat_len_A,0:sub_mat_len_A]

  A2=A[0:sub_mat_len_A,sub_mat_len_A:]

  A3=A[sub_mat_len_A:,0:sub_mat_len_A]

  A4=A[sub_mat_len_A:,sub_mat_len_A:]


  sub_mat_len_B=len(A)//2


  B1=B[0:sub_mat_len_B,0:sub_mat_len_B]

  B2=B[0:sub_mat_len_B,sub_mat_len_B:]

  B3=B[sub_mat_len_B:,0:sub_mat_len_B]

  B4=B[sub_mat_len_B:,sub_mat_len_B:]


  C1 = sub_mat(A1, B1) + sub_mat(A2, B3)
  C2 = sub_mat(A1, B2) + sub_mat(A2, B4)
  C3 = sub_mat(A3, B1) + sub_mat(A4, B3)
  C4 = sub_mat(A3, B2) + sub_mat(A4, B4)

  result = np.concatenate((np.concatenate((C1, C2), axis=1),np.concatenate((C3, C4), axis=1)), axis=0)
  return result

A = np.array([[1, 2, 3, 0], [4, 5, 6, 7], [7, 8, 9, 3], [1, 0, 3, 4]])
B = np.array([[3, 2, 1, 8], [9, 7, 8, 3], [6, 4, 5, 1], [6, 8, 1, 0]])

result =sub_mat(A, B)
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult:")
print(result)



```

```python
#fecth as many pages as we provide i.e we have to provide the no_of_pages we want to scrape form the amazon website

def scrape_amazon(search_keyword, num_pages):
    base_url = 'https://www.amazon.in/s'
    data = []

    for page in range(1, num_pages + 1):
        params = {'k': search_keyword, 'page': page}
        webpage = requests.get(base_url, params=params)
        soup = BeautifulSoup(webpage.content, 'html.parser')

        products = soup.find_all('div', class_='puis-card-container s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v1x9xtdrqwkf2b2sy3gv3bhvhpn s-latency-cf-section puis-card-border')

        for product in products:
            name = product.find('h2').text.strip()
            rating = product.find('span', class_='a-icon-alt').text.strip() if product.find('span', class_='a-icon-alt') else 'N/A'
            reviews = product.find('span', class_='a-size-base s-underline-text').text.strip() if product.find('span', class_='a-size-base s-underline-text') else 'N/A'
            price = product.find('span', class_='a-price-whole').text.strip() if product.find('span', class_='a-price-whole') else 'N/A'

            data.append({'product_name': name, 'rating': rating, 'reviews': reviews, 'price': price})
        time.sleep(2)
    return pd.DataFrame(data)

search_keyword = 'speaker'
num_pages_to_scrape = 50
df = scrape_amazon(search_keyword, num_pages_to_scrape)
print(df)
```
